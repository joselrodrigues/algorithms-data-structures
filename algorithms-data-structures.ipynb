## Common Big O Values

La notacin Big O se utiliza para analizar y comparar algoritmos basndose en su eficiencia en trminos de tiempo de ejecucin y espacio de memoria. Esta notacin permite identificar cmo el tiempo de ejecucin y el espacio de memoria de un algoritmo pueden aumentar en funcin del tamao de entrada 'n'. Es importante tener en cuenta que la notacin Big O es una estimacin del peor escenario posible, es decir, el tiempo mximo que podra tomar un algoritmo en funcin del tamao de entrada.

A continuacin, se muestran las complejidades ms comunes en la notacin Big O con una descripcin ms detallada:

O(1) - Constante: La complejidad del algoritmo no depende del tamao de entrada. Independientemente de cun grande o pequeo sea 'n', el tiempo de ejecucin y el espacio de memoria permanecen constantes.

O(n) - Lineal: La complejidad del algoritmo aumenta de manera proporcional al tamao de entrada. Si 'n' se duplica, el tiempo de ejecucin y el espacio de memoria tambin se duplican. Un ejemplo comn de complejidad O(n) es recorrer todos los elementos de una lista.

O(n^2) - Cuadrtico: La complejidad del algoritmo aumenta al cuadrado del tamao de entrada. Si 'n' se duplica, el tiempo de ejecucin aumenta cuatro veces. Un ejemplo comn de complejidad O(n^2) es el algoritmo de ordenacin por burbuja.

O(log n) - Logartmico: La complejidad del algoritmo aumenta logartmicamente con respecto al tamao de entrada. A medida que 'n' crece, el tiempo de ejecucin y el espacio de memoria aumentan de manera ms lenta. Un ejemplo comn de complejidad O(log n) es la bsqueda binaria.

O(n log n) - Lineal-Logartmica: La complejidad del algoritmo es una combinacin de crecimiento lineal y logartmico. Un ejemplo comn de complejidad O(n log n) es el algoritmo de ordenacin rpida (Quicksort).

### Frecuency counter

El patrn Frequency Counter (Contador de Frecuencias) es una tcnica de resolucin de problemas comnmente utilizada en algoritmos, especialmente en aquellos relacionados con la comparacin de elementos en estructuras de datos como arreglos o cadenas. Este enfoque implica contar la frecuencia de aparicin de elementos individuales en una estructura de datos y almacenar esos recuentos en una estructura de datos auxiliar, como un objeto o un mapa.

El objetivo principal del patrn Frequency Counter es evitar el uso de bucles anidados, lo que puede llevar a una complejidad de tiempo de ejecucin alta, como O(n^2). Al utilizar un contador de frecuencias, la complejidad del tiempo de ejecucin se reduce a menudo a O(n), lo que mejora significativamente la eficiencia del algoritmo.

````javascript
function same(arr1, arr2){
    if(arr1.length !== arr2.length){
        return false;
    }
    let frequencyCounter1 = {}
    let frequencyCounter2 = {}
    for(let val of arr1){
        frequencyCounter1[val] = (frequencyCounter1[val] || 0) + 1
    }
    for(let val of arr2){
        frequencyCounter2[val] = (frequencyCounter2[val] || 0) + 1        
    }
    console.log(frequencyCounter1);
    console.log(frequencyCounter2);
    for(let key in frequencyCounter1){
        if(!(key ** 2 in frequencyCounter2)){
            return false
        }
        if(frequencyCounter2[key ** 2] !== frequencyCounter1[key]){
            return false
        }
    }
    return true
}

same([1,2,3,2,5], [9,1,4,4,11])
````

La funcin same toma dos arreglos como entrada, arr1 y arr2, y devuelve true si cada nmero en arr1 tiene un cuadrado correspondiente en arr2, con la misma frecuencia. De lo contrario, devuelve false. La funcin utiliza el concepto de contadores de frecuencia para realizar esta tarea de manera eficiente.

Aqu est el paso a paso de lo que hace la funcin:

Comprueba si la longitud de arr1 y arr2 es diferente. Si es as, devuelve false, ya que no pueden tener nmeros cuadrados correspondientes en ese caso.
Crea dos objetos, frequencyCounter1 y frequencyCounter2, para almacenar la frecuencia de los nmeros en arr1 y sus cuadrados en arr2, respectivamente.
Itera a travs de arr1 y cuenta la frecuencia de cada nmero, almacenndola en frequencyCounter1.
Itera a travs de arr2 y cuenta la frecuencia de cada nmero, almacenndola en frequencyCounter2.
Imprime en consola frequencyCounter1 y frequencyCounter2 para mostrar las frecuencias.
Itera a travs de las claves en frequencyCounter1 y verifica si el cuadrado de la clave (key ** 2) est presente en frequencyCounter2. Si no lo est, devuelve false.
Si el cuadrado est presente en frequencyCounter2, verifica si la frecuencia del cuadrado en frequencyCounter2 es igual a la frecuencia del nmero original en frequencyCounter1. Si no es as, devuelve false.
Si todas las condiciones anteriores se cumplen, la funcin devuelve true.
En el ejemplo dado:


```bash

Esto se debe a que, aunque hay cuadrados correspondientes para algunos nmeros (por ejemplo, 1, 2 y 3), el nmero 5 en arr1 no tiene un cuadrado correspondiente en arr2, y el nmero 11 en arr2 no tiene una raz cuadrada en arr1. Adems, la frecuencia de los nmeros y sus cuadrados no coincide.

### Multiple pointers 

El patrn Multiple Pointers (Mltiples Punteros) es otra tcnica de resolucin de problemas utilizada en algoritmos, particularmente en aquellos que involucran estructuras de datos lineales como arreglos o cadenas. Este enfoque implica el uso de dos o ms punteros, o variables que apuntan a diferentes posiciones dentro de la estructura de datos, para realizar comparaciones, bsquedas o clculos sin necesidad de bucles anidados.

El objetivo principal del patrn Multiple Pointers es reducir la complejidad del tiempo de ejecucin al evitar bucles anidados y, en su lugar, realizar operaciones utilizando punteros adicionales. Esto a menudo mejora la eficiencia del algoritmo, reduciendo la complejidad del tiempo de ejecucin a O(n) o incluso O(log n) en algunos caso

````javascript
function countUniqueValues(arr){
    if(arr.length === 0) return 0;
    var i = 0;
    for(var j = 1; j < arr.length; j++){
        if(arr[i] !== arr[j]){
            i++;
            arr[i] = arr[j]
        }
    }
    return i + 1;
}
countUniqueValues([1,2,2,5,7,7,99])
````

La funcin countUniqueValues toma un arreglo ordenado arr como entrada y devuelve la cantidad de valores nicos presentes en el arreglo. La funcin utiliza el patrn de Mltiples Punteros para lograr esto de manera eficiente.

Aqu est el paso a paso de lo que hace la funcin:

Si la longitud del arreglo es 0, devuelve 0, ya que no hay valores nicos en un arreglo vaco.
Crea dos punteros, i y j. Inicializa i en 0 y j en 1. Estos punteros se usarn para comparar los elementos adyacentes en el arreglo.
Comienza un bucle for que recorre el arreglo desde el segundo elemento hasta el ltimo elemento (ndice j).
Dentro del bucle, compara los elementos en las posiciones i y j. Si los elementos son diferentes (arr[i] !== arr[j]), significa que se ha encontrado un nuevo valor nico.
Si se encuentra un nuevo valor nico, incrementa i en 1 e iguala el elemento en la posicin i al elemento en la posicin j. Esto asegura que el arreglo est actualizado con los valores nicos encontrados hasta ahora.
Si los elementos en las posiciones i y j son iguales, el bucle contina con el siguiente ndice j.
Repite los pasos 4-6 hasta que el bucle termine.
Devuelve i + 1, que representa la cantidad de valores nicos en el arreglo.
En el ejemplo proporcionado:
```bash
countUniqueValues([1, 2, 2, 5, 7, 7, 99])
```
La funcin devuelve 5, ya que hay 5 valores nicos en el arreglo: 1, 2, 5, 7 y 99. Al usar el patrn de Mltiples Punteros, la funcin evita bucles anidados y resuelve el problema con una complejidad de tiempo de ejecucin de O(n).

### Sliding Window

El patrn Sliding Window (Ventana Deslizante) es una tcnica de resolucin de problemas utilizada en algoritmos que involucran estructuras de datos lineales, como arreglos o cadenas. Este enfoque consiste en crear una "ventana" o subconjunto de elementos dentro de la estructura de datos y desplazarla a lo largo de la estructura mientras se realizan clculos o comparaciones en los elementos contenidos en la ventana. La ventana puede ser de tamao fijo o variable, dependiendo del problema en cuestin.

El objetivo principal del patrn Sliding Window es mejorar la eficiencia del algoritmo al evitar la necesidad de recalcular ciertos valores o repetir operaciones en cada paso del proceso. En lugar de utilizar bucles anidados, que pueden llevar a una alta complejidad en el tiempo de ejecucin (como O(n^2)), el patrn Sliding Window a menudo reduce la complejidad a O(n).

````javascript
function maxSubarraySum(arr, num){
  let maxSum = 0;
  let tempSum = 0;
  if (arr.length < num) return null;
  for (let i = 0; i < num; i++) {
    maxSum += arr[i];
  }
  tempSum = maxSum;
  for (let i = num; i < arr.length; i++) {
    tempSum = tempSum - arr[i - num] + arr[i];
    maxSum = Math.max(maxSum, tempSum);
  }
  return maxSum;
}

maxSubarraySum([2,6,9,2,1,8,5,6,3],3)
````

La funcin maxSubarraySum toma un arreglo de nmeros arr y un nmero entero num como argumentos, y devuelve la suma mxima de num elementos consecutivos en el arreglo. La funcin utiliza el patrn Sliding Window (Ventana Deslizante) para lograr esto de manera eficiente.

Aqu est el paso a paso de lo que hace la funcin:

Inicializa las variables maxSum y tempSum en 0. maxSum almacenar la suma mxima encontrada hasta ahora, mientras que tempSum almacenar la suma de los elementos en la ventana actual.
Si la longitud del arreglo arr es menor que num, no hay suficientes elementos para calcular una subsuma, por lo que la funcin devuelve null.
Utiliza un bucle for para calcular la suma de los primeros num elementos del arreglo y asigna este valor tanto a maxSum como a tempSum.
Comienza otro bucle for que recorre el arreglo desde el ndice num hasta el final del arreglo.
En cada iteracin del bucle, actualiza tempSum restando el primer elemento de la ventana actual (arr[i - num]) y sumando el siguiente elemento fuera de la ventana (arr[i]). Esto desliza la ventana hacia adelante.
Compara tempSum con maxSum y actualiza maxSum si tempSum es mayor, utilizando la funcin Math.max().
Repite los pasos 5 y 6 hasta que el bucle termine.
Devuelve maxSum, que representa la suma mxima de num elementos consecutivos en el arreglo.
En el ejemplo proporcionado:

```bash
maxSubarraySum([2, 6, 9, 2, 1, 8, 5, 6, 3], 3)
```
La funcin devuelve 19, ya que la suma mxima de 3 elementos consecutivos en el arreglo es 9 + 2 + 8 = 19. Al utilizar el patrn Sliding Window, la funcin evita bucles anidados y resuelve el problema con una complejidad de tiempo de ejecucin de O(n).
